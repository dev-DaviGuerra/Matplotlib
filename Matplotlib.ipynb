```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x ** 2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y, 'red')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Título')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g--')
```

## Orientação à objetos no matplotlib

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'blue')
axes.set_xlabel('Label x')
axes.set_ylabel('Label y')
axes.set_title('Título')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y, 'blue')
axes.set_xlabel('Label x')
axes.set_ylabel('Label y')
axes.set_title('Título')

axes2.plot(y, x, 'red')
axes2.set_xlabel('Label x2')
axes2.set_ylabel('Label y2')
axes2.set_title('Título2')
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, 'r--')
ax.set_title('Título do nosso subplot')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(16, 8))

for axis in ax:
  axis.plot(x, y, 'r')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, y, 'r')
ax[1].plot(y, x, 'g--')
```
